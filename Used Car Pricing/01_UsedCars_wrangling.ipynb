#### Handle Missing Values in "normalized-losses"

1. Convert the `"normalized-losses"` column to numeric values.  
2. Calculate the mean value of the column.  
3. Replace all missing values (`NaN`) in `"normalized-losses"` with the calculated mean.  


```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
df["normalized-losses"] = df["normalized-losses"].replace(np.nan, avg_norm_loss)
```

### Handle Missing Values in "bore"

1. Convert the `"bore"` column to numeric values.  
2. Calculate the mean value of the column.  
3. Replace all missing values (`NaN`) in `"bore"` with the calculated mean.  


```python
avg_bore = df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
df["bore"] = df["bore"].replace(np.nan, avg_bore)
```

### Handle Missing Values in "stroke"

```python
# Calculate the mean value for "stroke" column
avg_stroke = df["stroke"].astype("float").mean(axis=0)
print("Average of stroke:", avg_stroke)

# Replace NaN by mean value in "stroke" column
df["stroke"] = df["stroke"].replace(np.nan, avg_stroke)
```

### Handle Missing Values in "peak-rpm"

```python
# Calculate the mean value for "peak-rpm" column
avg_peakrpm = df["peak-rpm"].astype("float").mean(axis=0)
print("Average of peak-rpm:", avg_peakrpm)

# Replace NaN by mean value in "peak-rpm" column
df["peak-rpm"] = df["peak-rpm"].replace(np.nan, avg_peakrpm)
```

### Handle "num of doors" categorical values

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
#idxmax() gives you the index (category) with the highest count.
```

```python
#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-dors"]=df["num-of-doors"].replace(np.nan, "four")
```

```python
# Or check counts again
print(df["num-of-doors"].value_counts())
```

### Dropping rows with missing "Price" values and resetting the index


```python
# Price (drop rows with NaN, then reset index)
df = df.dropna(subset=["price"], axis=0)
df = df.reset_index(drop=True)
```

####  Summary of Strategy

* <b>Numeric features</b> â†’ Replace NaN with mean (keeps them usable).

* <b>Categorical features</b> â†’ Replace NaN with mode (most frequent category).

* <b>Target variable (price)</b> â†’ Drop rows with missing values (donâ€™t risk bias).

### Example of Binning data

### Binning
#### Why binning?

* Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis.

#### Example:

* In your data set, "horsepower" is a real valued variable ranging from 48 to 288 and it has 59 unique values. What if you only care about * the price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)? You can rearrange them into * three â€˜bins' to simplify analysis.

* Use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins.

```python
# Convert to float first (floats can store NaN)
df["horsepower"] = df["horsepower"].astype(float)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(df["horsepower"])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
# Create a new column 'horsepower-binned' by segmenting 'horsepower' into bins
df['horsepower-binned'] = pd.cut(
    df['horsepower'],        # the original horsepower column
    bins,                    # the array of bin edges created using np.linspace
    labels=group_names,      # assign a label to each bin (e.g., 'Low', 'Medium', 'High')
    include_lowest=True      # include the minimum value in the first bin
)

# Display the first 20 rows of the original horsepower and the new binned column
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
#import matplotlib.pyplot as plt

plt.bar(group_names, df["horsepower-binned"].value_counts().sort_index())
plt.xlabel("Horsepower Bins")
plt.ylabel("Count")
plt.title("Distribution of Horsepower Bins")
plt.show()
```

```python
%matplotlib inline
#import matplotlib.pyplot as plt
# Draw histogram of 'horsepower' with 3 bins
plt.hist(df["horsepower"], bins=3)

# Set x and y labels and plot title
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Bins")

# Display the plot
plt.show()
```
